## 3. Keras OCR

```python
#!pip install keras-ocr
```

```python
import keras_ocr
```

```python
import matplotlib.pyplot as plt
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
images = [keras_ocr.tools.read('image.jpg')]
```

```python
prediction_groups = pipeline.recognize(images)
```

```python
# prediction_groups # this outputs the result together with the coordinates of boxes
```

```python
for group in prediction_groups[0]:
    print(group[0])
```

So far this did not make much sense, obviously this is not a German model which tries to read a German text here
